### 연령별 분석

```python
df['birthdate'] = pd.to_datetime(df['birthdate'])
df_active['birthdate'] = pd.to_datetime(df_active['birthdate'])
df_inactive['birthdate'] = pd.to_datetime(df_inactive['birthdate'])

print(df['birthdate'].dtypes)
print(df_active['birthdate'].dtypes)
print(df_inactive['birthdate'].dtypes)
```

```python
# 나이 계산
current_year = datetime.now().year

df['year_of_birth'] = df['birthdate'].dt.year
df_active['year_of_birth'] = df_active['birthdate'].dt.year
df_inactive['year_of_birth'] = df_inactive['birthdate'].dt.year

df['age'] = current_year - df['year_of_birth']
df_active['age'] = current_year - df_active['year_of_birth']
df_inactive['age'] = current_year - df_inactive['year_of_birth']
```

```python
from pandas.api.types import CategoricalDtype

# 정렬을 원하는 순서로 카테고리를 지정합니다.
cat_type = CategoricalDtype(categories=['10대 이하','10대','20대', '30대', '40대', '50대' , '60대', '70대이상'], ordered=True)

# 연령 그룹 생성
bins = [0, 10, 20, 30, 40, 50, 60, 70, np.inf]
labels = ['10대 이하', '10대', '20대', '30대', '40대', '50대', '60대', '70대이상']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels).astype(cat_type)
df_active['age_group'] = pd.cut(df_active['age'], bins=bins, labels=labels).astype(cat_type)
df_inactive['age_group'] = pd.cut(df_inactive['age'], bins=bins, labels=labels).astype(cat_type)


```

```python
age_groups = df['age_group'].sort_values().unique()
age_groups
```

```python
# 전체 고객 연령대별 인원 수
for i in age_groups:
    customer_i = df[df['age_group'] == i]['customer_id'].nunique()
    print(f'{i} : {customer_i}')
```

```python
for i in age_groups:
    total_customer_i = df_active[df_active['age_group'] == i]['customer_id'].nunique()
    print(f'{i} : {total_customer_i}')
```

```python
for i in age_groups:
    total_customer_i = df[df['age_group'] == i]['customer_id'].nunique()
    success_customer_i = df[((df['payment_status'] == 'Success')) & (df['age_group'] == i)]['customer_id'].nunique()
    bounce_rate_i = (1 - (success_customer_i / total_customer_i)) * 100
    print(f'이탈률 - {i} : {bounce_rate_i}')
```
